```
# @author: Nº 30000794 Nome: Ricardo Clemente Curso: Informática de Gestão

import sklearn.datasets as datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.preprocessing import StandardScaler

def aggregation(data):
  
  new_data =np.concatenate((data.data, data.target.reshape(-1,1)*100000), axis=1)
  names = data.feature_names + ["Preços"]
  
  dframe = pd.DataFrame(data=new_data, columns=names)
  
  
  return dframe.aggregate(["sum", "mean", "max"])
  
  
def sampling(data,nr_samples):
  
  new_data =np.concatenate((data.data, data.target.reshape(-1,1)*100000), axis=1)
  names = data.feature_names + ["Preços"]
  
  dframe = pd.DataFrame(data=new_data, columns=names)
  
  
  return dframe.sample(nr_samples)


def dimentional_reduction(data):
  
  
  new_data =np.concatenate((data.data, data.target.reshape(-1,1)*100000), axis=1)
  names = data.feature_names + ["Preços"]
  
  dframe = pd.DataFrame(data=new_data, columns=names)
  X = dframe.ix[:,0:9].values
  y = dframe.ix[:,4].values
  X_std = StandardScaler().fit_transform(X)
  mean_vec = np.mean(X_std, axis=0)
  cov_mat = (X_std - mean_vec).T.dot((X_std - mean_vec)) / (X_std.shape[0]-1)
  cov_mat = np.cov(X_std.T)

  eig_vals, eig_vecs = np.linalg.eig(cov_mat)
  print('\nEigenvalues \n%s' %eig_vals)
  tot = sum(eig_vals)
  var_exp = [ (i / tot)*100 for i in eig_vals ]
  
  
  print("vectors\n",eig_vecs)
  return "PCA:",var_exp


def feature_subset_selection(data):
  
  
  new_data =np.concatenate((data.data, data.target.reshape(-1,1)*100000), axis=1)
  names = data.feature_names + ["Preços"]
  
  dframe = pd.DataFrame(data=new_data, columns=names)
  
  return dframe.loc[0:10,["MedInc","AveRooms","AveBedrms","Latitude","Longitude","Preços"]]
  
  
def feature_creation(data):
  
  
  new_data =np.concatenate((data.data, data.target.reshape(-1,1)*100000), axis=1)
  names = data.feature_names + ["Preços"]
  dframe = pd.DataFrame(data=new_data, columns=names)
  x=dframe.Preços>300000
  
  dframe["CasaCara"]=x
  
  
  return dframe.head(20)


def bin(data):

  new_data =np.concatenate((data.data, data.target.reshape(-1,1)*100000), axis=1)
  names = data.feature_names + ["Preços"]
  dframe = pd.DataFrame(data=new_data, columns=names)
  x = dframe.Preços[0:2]
  return pd.get_dummies(x)


def atribute_transf(data):
  
  new_data =np.concatenate((data.data, data.target.reshape(-1,1)*100000), axis=1)
  names = data.feature_names + ["Preços"]
  dframe = pd.DataFrame(data=new_data, columns=names)
  
  dframe["AvgRoom_Without_AvgBedRooms"]=dframe.AveRooms-dframe.AveBedrms
  
  return dframe.head(10)
  
```

```
data = datasets.california_housing.fetch_california_housing()
aggregation(data)
```

```
sampling(data,10)
```

```
dimentional_reduction(data)
```

```
feature_subset_selection(data)
```

```
feature_creation(data)
```

```
bin(data)
```

```
atribute_transf(data)
```
