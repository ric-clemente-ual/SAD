No nosso trabalho o  Precision e Recall são ambos importantes , em que talvez um medidor como F1-score seja o mais apropriado para o nosso trabalho em que nos da uma média entre os valores da Precision e o Recall, em que quanto mais o valor da media se aproximar de 1 melhor será a prestação de cada label que foi classificada.

Entre o Recall e o Precision , o Recall é o mais relevante em que nos da % de resultados  positivos correctos de uma label(TPositives) em relação a amostra global de resultados que deviam ser considerados como positivos dessa label(TPositives + FNegatives).
